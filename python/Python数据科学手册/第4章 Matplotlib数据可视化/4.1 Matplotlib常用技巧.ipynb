# 4.1.3 用不用show()？如何显示图形

## 1. 在脚本中画图

在脚本中使用Matplotlib，那么显示图形的时候必须使用plt.show()。

## 2. 在Ipython shell中画图

需要在启动ipython后使用%matplotlib魔法命令。

## 3. 在Ipython Notebook中画图

与Ipython shell类似，需要使用%matplotlib命令。有两种表现形式。

- %matplotlib notebook 会在Notebook中启动**交互式**图形。
- %matplotlib inline会在Notebook中启动**静态**图形。

```python
import numpy as np
x = np.linspace(0, 10, 100)
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
```
