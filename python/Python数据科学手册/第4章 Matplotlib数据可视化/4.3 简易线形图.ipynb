```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
fig = plt.figure()
ax = plt.axes()
```

figure可以被看成是一个可以容纳各种坐标轴、图形、文字和标签的容器。axes是一个带有刻度和标签的矩形，最终会包含所有可视化的图形元素。  
变量fig表示一个图形实例，变量ax表示一个坐标轴实例或一组坐标轴实例。

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
```

也可以使用MATLAB风格的接口

```python
plt.plot(x, np.sin(x))
```

在一张图中创建多条线，可以重复调用plot命令

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

# 4.3.1 调整图形：线条的颜色与风格

```python
plt.plot(x, np.sin(x - 0), color='blue')
plt.plot(x, np.sin(x - 1), color='g')
plt.plot(x, np.sin(x - 2), color='0.75')
plt.plot(x, np.sin(x - 3), color='#FFDD44')
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3))
plt.plot(x, np.sin(x - 5), color='chartreuse')
```

可以用linestyle调整线条的风格

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')
```

简写形式

```python
plt.plot(x, x + 0, linestyle='-') # 实线
plt.plot(x, x + 1, linestyle='--') # 虚线
plt.plot(x, x + 2, linestyle='-.') # 点划线
plt.plot(x, x + 3, linestyle=':') # 点实线
```

```python
plt.plot(x, x + 0, '-g') # 绿色实线
plt.plot(x, x + 1, '--c') # 青色虚线
plt.plot(x, x + 2, '-.k') # 黑色点划线
plt.plot(x, x + 3, ':r') # 红色实点线
```

# 4.3.2 调整图形：坐标轴上下限

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

```python
plt.plot(x, np.sin(x))

# 逆序
plt.xlim(10, 0)
plt.ylim(1.2, -1.2)
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]) # [xmin, xmax, ymin, ymax]
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight') # 按照图形内容自动收紧坐标轴，不留空白区域
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
```

# 4.3.3 设置图形标签

```python
plt.plot(x, np.sin(x))
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

设置图例

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend()
```
