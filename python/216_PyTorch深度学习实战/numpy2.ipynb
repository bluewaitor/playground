```python
from PIL import Image
im = Image = Image.open('test.jpeg')
```

```python
im.size
```

```python
import numpy as np
im_pillow = np.asarray(im)

im_pillow.shape
```

```python
r = im_pillow[:, :, 0]
g = im_pillow[:, :, 1]
b = im_pillow[:, :, 2]
```

```python
zeros = np.zeros((im_pillow.shape[0], im_pillow.shape[1], 2))
zeros.shape
```

```python
r1 = r[:, :, np.newaxis]
g1 = r[:, :, np.newaxis]
b1 = r[:, :, np.newaxis]
```

```python
r1.shape
```

```python
r2 = np.concatenate((r1, zeros), axis=2)
r2.shape
```

```python
g2 = np.zeros(im_pillow.shape)
g2[:, :, 1] = g1[:, :, 0]
```

```python
b2 = np.zeros(im_pillow.shape)
b2[:, :, 2] = b1[:, :, 0]
```

```python
from matplotlib import pyplot as plt
plt.subplot(2, 2, 1)
plt.title('Origin Image')
plt.imshow(im_pillow)
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title('Red Channel')
plt.imshow(r2.astype(np.uint8))
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title('Green Channel')
plt.imshow(g2.astype(np.uint8))
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title('Blue Channel')
plt.imshow(b2.astype(np.uint8))
plt.axis('off')

plt.savefig('./rgb_pillow.png', dpi=150)
```
